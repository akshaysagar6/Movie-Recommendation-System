### Now Convert Genres

```python
df.iloc[0].genres #this is a string of list of distionary so we need to convert it into a list
```

```python
# to convert this into a list we can use a module  ast .literal_eval
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
       l.append(i['name']) 
    return l
```

```python
df.genres.apply(convert)
```

```python
df.genres=df.genres.apply(convert)
```

### Taking only top 3 actors

```python
def convert_cast(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
df.cast.apply(convert_cast)
```

```python
df.cast=df.cast.apply(convert_cast)
df
```

```python
df.loc[0].cast
```

### Now find the director

```python
df.loc[0].crew
```

```python
def crew(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
df.crew.apply(crew)
```

```python
df.crew=df.crew.apply(crew)
df
```

### here we need to apply stemming

### Apply stemming

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
df_new['tags']=df_new['tags'].apply(stem)
```

```python
df_new
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(df_new['tags']).toarray()
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_sim=cosine_similarity(vectors)
movie_sim
```

```python
np.fill_diagonal(movie_sim,0)
movie_sim
```

```python
movie_sim.shape
```

```python
def recommed(movie):
    movie_index=df_new[df_new['title']==movie].index[0]#to get the index of movies
    distance=movie_sim[movie_index].argsort()[:-11:-1]#find the index of top similar movies from sovie_sim matrix, #gives top 5 movie index
   
    movie_recom=df_new.loc[distance]
    print(movie_recom['title'])
    
```

```python
recommed('Avatar')
```
